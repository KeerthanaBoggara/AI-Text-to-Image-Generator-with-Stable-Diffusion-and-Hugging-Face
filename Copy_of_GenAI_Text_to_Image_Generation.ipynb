```python
import streamlit as st
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image, ImageDraw, ImageFont
from io import BytesIO
from deep_translator import GoogleTranslator
import imageio

# Configure Streamlit
st.set_page_config(page_title="Text-to-Image Generator", layout="centered")
```

```python
# App Title
st.title("AI Text-to-Image Generator")
st.markdown("Enter a prompt and generate an image using Stable Diffusion")
```

```python
#Sidebar Controls
with st.sidebar:
    st.header("Settings")

    model_choice = st.selectbox("Choose a model:", [
        "runwayml/stable-diffusion-v1-5",
        "stabilityai/stable-diffusion-2-1"
    ])

    resolution = st.selectbox("Image Resolution", ["512x512", "768x768", "1024x1024"])
    width, height = map(int, resolution.split("x"))

    lang = st.selectbox("Prompt Language", ["English", "French", "Spanish", "German", "Tamil", "Japanese"])

    guidance_scale = st.slider("Guidance Scale", 1.0, 20.0, 7.5, 0.5)
    num_inference_steps = st.slider("Inference Steps", 10, 100, 50, 10)
```

```python
# Prompt Input & Session State
prompt = st.text_input("Enter your prompt:", placeholder="e.g. A futuristic city in the clouds at sunset")

if "history" not in st.session_state:
    st.session_state.history = []

if "gallery" not in st.session_state:
    st.session_state.gallery = []

# Sidebar prompt history
with st.sidebar:
    st.markdown("### Prompt History")
    for p in reversed(st.session_state.history[-5:]):
        st.markdown(f"- {p}")
```

```python
#Load Stable Diffusion Model
@st.cache_resource
def load_pipeline(model_id):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    pipe = StableDiffusionPipeline.from_pretrained(
        model_id,
        torch_dtype=torch.float16 if device == "cuda" else torch.float32,
    )
    pipe.to(device)
    return pipe

pipe = load_pipeline(model_choice)
```

```python
# Language Translation Helper
def translate_prompt(text, target="en"):
    if lang != "English":
        return GoogleTranslator(source='auto', target=target).translate(text)
    return text
```

```python
# Watermark Helper
def add_watermark(image, text="© Keerthana AI"):
    watermark = image.copy()
    draw = ImageDraw.Draw(watermark)
    font = ImageFont.load_default()
    position = (10, watermark.height - 20)
    draw.text(position, text, font=font, fill="white")
    return watermark
```

```python
# Image Generation Block
if st.button("Generate Image"):
    if prompt:
        st.session_state.history.append(prompt)
        with st.spinner("Generating image..."):
            translated = translate_prompt(prompt)
            image = pipe(
                translated,
                guidance_scale=guidance_scale,
                num_inference_steps=num_inference_steps,
                height=height,
                width=width
            ).images[0]
            watermarked = add_watermark(image)
            st.session_state.gallery.append(watermarked)

            st.image(watermarked, caption=" Generated Image", use_column_width=True)

            buf = BytesIO()
            watermarked.save(buf, format="PNG")
            byte_im = buf.getvalue()
            st.download_button("Download Image", byte_im, file_name="generated_image.png")
    else:
        st.warning("Please enter a prompt to generate an image.")
```

```python
# Gallery Display
if st.session_state.gallery:
    st.markdown("### Image Gallery (Recent 5)")
    for img in reversed(st.session_state.gallery[-5:]):
        st.image(img, use_container_width=True)
```
